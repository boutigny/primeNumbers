```python
img = Image.open("Joconde.jpg")
width, height = img.size
print(width, height)
img = img.crop((0, 0, width, height-30)).resize((33,49))
```

```python
img
```

```python
im = img.convert("P", palette = Image.ADAPTIVE, colors = 10)
```

```python
arr = np.array(im)
```

```python
pal = im.getpalette()
```

```python
num = arr.ravel()
```

```python
num2 = ''.join(map(str, num))
```

```python
num2
```

```python
prim = gmpy2.next_prime(mpz(num2)-5000)
```

```python
prim
```

```python
str(prim)
```

```python
arr2 = np.array(list(str(prim)))
```

```python
arr3 = arr2.reshape((49,33))
```

```python
arr3
```

```python
im_new = Image.fromarray(arr3.astype("uint8"), "P")
im_new.putpalette(pal)
```

```python
im_new
```

```python
im_new.save('new3.bmp')
```
